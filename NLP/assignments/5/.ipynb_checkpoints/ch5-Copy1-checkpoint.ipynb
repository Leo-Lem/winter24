## Gender identification 
Names ending in <span style="color:blue">a, e and i</span> are likely to be female, while names ending in <span style="color:blue">k, o, r, s</span> and <span style="color:blue">t</span> are likely to be male. The following is the <span style="color:blue">feature extractor</span>, which returns a <span style="color:blue">feature set</span> that maps feature names to their values.

```python
#feature extraction
def gender_features(word):
    return {'last_letter': word[-1]}
gender_features('Shrek')
```

### Train a model with list of names

```python
import nltk
nltk.download('names')
```

```python
from nltk.corpus import names
names.words('female.txt')
```

```python
import nltk
from nltk.corpus import names
# male and female names
labeled_names = ([(name, 'male') for name in names.words('male.txt')] + 
                 [(name, 'female') for name in names.words('female.txt')])
import random
# shuffle the names (in place: so the shuffled values in labled_names)
random.shuffle(labeled_names)
#Extracting features
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]
#print the first 10 features
print("Feature Example",featuresets[0:10] )
# Split into ttrain and test sets
train_set, test_set = featuresets[500:], featuresets[:500]
print ("Training =",len(train_set), "Test =",len(test_set))
# train a classifier with Naive Bayes
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

### Testing the classifier

```python
print("Neon is a ", classifier.classify(gender_features('Neo')), " name")
print ("Trinity is a ", classifier.classify(gender_features('Trinity'))," name")
```

### Evaluate the classifier on larger dataset

```python
print(nltk.classify.accuracy(classifier, test_set))
```

# Document Classification in NLTK
Lets see how to build a document classifier using the <span style="color:blue">Movie Reviews Corpus</span>, where the task is to categorizes each review as positive or negative.

```python
from nltk.corpus import movie_reviews
# documents as[Document, lable]
documents = [(list(movie_reviews.words(fileid)), category)
              for category in movie_reviews.categories()
              for fileid in movie_reviews.fileids(category)]
# Shuffle documents so that we similar documents will not be only in training/test sets
random.shuffle(documents)
```

```python
# movie_reviews.fileids
```

```python
# how the first document looks like ==> (list(tokens), label)
documents[0]
```

```python
# Show the lable of the first document
documents[0][1]
```

## Feature extractor
A simple option to determine the sentiment of the document is to check the <span style="color:blue">presence of words in the document as a feature</span> (a boolean feature). We can list the most frequent words first, for example the most common 2k words and see if the document contains those words.

```python
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
# first 2k words
word_features = list(all_words)[:2000] 
def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features
```

```python
# generate features for the document with fieleid 'pos/cv957_8737.txt'
print(document_features(movie_reviews.words('pos/cv957_8737.txt')))
```

## Train the classifier
We can use Naive Bayes classifier based on these features (presence and absence of words in the document)

```python
featuresets = [(document_features(d), c) for (d,c) in documents]
train_set, test_set = featuresets[100:], featuresets[:100]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

## Test the classifier
We can test the classifier on the test set and show the most informative 5 features

```python
#print the accuracy of the model on the test set
print(nltk.classify.accuracy(classifier, test_set))
classifier.show_most_informative_features(5)
```

# Scikit-Learn's Estimator API
The sklearn estimator API is developed where consistency is given a due regard. If you prepare your data properly, you can use any of the existing API without modifying your dataset. Most of the time, the default parameters are also designed to perform optimally.

The followings are the steps required when building sklearn models:

1. Choose a class of model by importing the appropriate <span style="color:blue">estimator class</span> from Scikit-Learn.
1. Choose <span style="color:blue">model hyperparameters</span> by instantiating this class with desired values.
1. Arrange data into a <span style="color:blue">features matrix and target vector</span> following the discussion above.
1. Fit the model to the data by calling the <span style="color:blue">fit()</span> method of the model instance.
1. Apply the model to new data, for supervised learning, often we predict labels for unknown data using the <span style="color:blue">predict()</span> method.

# Train POS tagger with scikit-learn

The files under <span style="color:blue">pos_tags</span> folder are training and test files, which are <span style="color:blue">TAB separated</span> with the token and its POS tag. Lets train a POS tagger, which include features such as the <span style="color:red">word itself</span> , <span style="color:red">two</span>  and  <span style="color:red">three character suffixes</span> , and <span style="color:red">previous word</span>  and so on.

```python
# Define features usefull for POS tagging
# Features: Word itself, 2 and 3 letter suffixes, previous and next word
# From here: https://nlpforhackers.io/training-pos-tagger/
def features(sentence, index):
    """ sentence: [w1, w2, ...], index: the index of the word """
    return {
        'word': sentence[index],
        'is_first': index == 0,
        'is_last': index == len(sentence) - 1,
        'is_capitalized': sentence[index][0].upper() == sentence[index][0],
        'is_all_caps': sentence[index].upper() == sentence[index],
        'is_all_lower': sentence[index].lower() == sentence[index],
        'prefix-1': sentence[index][0],
        'prefix-2': sentence[index][:2],
        'prefix-3': sentence[index][:3],
        'suffix-1': sentence[index][-1],
        'suffix-2': sentence[index][-2:],
        'suffix-3': sentence[index][-3:],
        'prev_word': '' if index == 0 else sentence[index - 1],
        'next_word': '' if index == len(sentence) - 1 else sentence[index + 1],
        'has_hyphen': '-' in sentence[index],
        'is_numeric': sentence[index].isdigit(),
        'capitals_inside': sentence[index][1:].lower() != sentence[index][1:]
    }
import pprint
# Show how the extracted POS features looks like
pprint.pprint(features(['This', 'is', 'a', 'sentence'], 0))
```

## Preparing the data 

Lets read TAB separated file, where a sentence is indicated with a  <span style="color:blue">new line</span>. We read each line, append to make a sentence until empty line is obtained. Then generate the features for each words using the method "features".

```python
def transform_file_to_dataset(filename):
    X, y = [], []
    with open(filename) as file:
        sent = []
        poses = []
        for line in file:
            if not line.strip(): 
                for index in range(len(sent)):
                    X.append(features(sent, index))
                    y.append(poses[index])
                sent = []
                poses = []
            else:
                lines = line.split("\t")
                sent.append(lines[0])
                poses.append(lines[1])
    return X, y
```

```python
pos_features, tags = transform_file_to_dataset("pos_data/wsj_pos.dev")
```

```python
print(pos_features[0],"\n===\n", tags[0])
```

```python
len(tags), len(pos_features)
```

## Training the model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.pipeline import Pipeline

# Use the DictVecorizer method, which transform lists of features-value mappings to vectors
print("Feature vectorizing started...")
dictVec = DictVectorizer(sparse=False)
pos_features_dictVec = dictVec.fit_transform(pos_features[:10000])
print("Feature vectorizing completed")

print("Training started...")
model = DecisionTreeClassifier(criterion='entropy')
model.fit(pos_features_dictVec, tags[:10000])   # Use only the first 10K samples)
print ('Training completed')
pos_features_test, tags_test = transform_file_to_dataset("pos_data/wsj_pos.test")

# Here we do not fit, we use the dictionary from the training and transform them
pos_features_test_dictv =dictVec.transform(pos_features_test)
# Compute the accuracy
print ("Accuracy:", model.score(pos_features_test_dictv, tags_test))
```

```python
# confusion matrix ploting --> https://coderzcolumn.com/tutorials/machine-learning/scikit-learn-sklearn-naive-bayes
from sklearn.metrics import confusion_matrix
def plot_confusion_matrix(Y_test, Y_preds):
    conf_mat = confusion_matrix(Y_test, Y_preds)
    print(conf_mat)
    fig = plt.figure(figsize=(4,4))
    plt.matshow(conf_mat, cmap=plt.cm.Blues, fignum=1)
    plt.yticks(range(2), range(2))
    plt.xticks(range(2), range(2))
    plt.colorbar();
    for i in range(2):
        for j in range(2):
            plt.text(i-0.2,j+0.1, str(conf_mat[j, i]), color='tab:red')
```

## Sklearn <span style="color:blue">Pipeline</span>
Sklear Pipeline is used to sequentially apply a list of <span style="color:blue">transforms</span> and a final <span style="color:blue">estimator.</span>  You can also use the Pipeline to run different estimators on group of transformations for easy performance evaluation and parameter tuning.

```python
# In stead of doing the transformation and training 
pipeline = Pipeline([
    ('vectorizer', DictVectorizer(sparse=False)),
    ('classifier', DecisionTreeClassifier(criterion='entropy'))
])
print ('Running the pipeline...')
pipeline.fit(pos_features[:10000], tags[:10000])  
print ('Pipleine completed')

print ("Accuracy:", pipeline.score(pos_features_test, tags_test))
```

## More examples on pipelines

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import joblib
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn import svm
from sklearn import tree
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Construct some pipelines
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('model', LogisticRegression(random_state=42))])

pipe_svm = Pipeline([('scl', StandardScaler()),
                     ('pca', PCA(n_components=2)),
                     ('model', svm.SVC(random_state=42))])

pipe_dt = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('model', tree.DecisionTreeClassifier(random_state=42))])

pipe_rg = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('model', Ridge())])

pipe_rg = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('model', Ridge())])

pipe_kn = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('model', KNeighborsClassifier(n_neighbors=2))])


# List of pipelines for ease of iteration
pipelines = [pipe_lr, pipe_svm, pipe_dt, pipe_rg, pipe_kn]

# Dictionary of pipelines and classifier types for ease of reference
pipe_dict = {0: 'Logistic Regression', 1: 'Support Vector Machine', 2: 'Decision Tree', 3: "Ridge", 4:"KNeighborsClassifier"}

# Fit the pipelines
for pipe in pipelines:
    pipe.fit(X_train, y_train)

# Compare accuracies
for idx, val in enumerate(pipelines):
    print('***** %s pipeline test accuracy*****: %.3f' % (pipe_dict[idx], val.score(X_test, y_test)))

# Identify the most accurate model on test data
best_acc = 0.0
best_model = 0
best_pipe = ''
for idx, val in enumerate(pipelines):
    if val.score(X_test, y_test) > best_acc:
        best_acc = val.score(X_test, y_test)
        best_pipe = val
        best_model = idx
        best_clf = idx
print ("best accuracy", best_acc)
print('Classifier with best accuracy: %s' % pipe_dict[best_model])

# Save pipeline to file
joblib.dump(best_pipe, 'best_pipeline.pkl', compress=1)
print('Saved %s pipeline to file' % pipe_dict[best_model])
```
